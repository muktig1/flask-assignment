```python
from flask import Flask, render_template, request, session, redirect, url_for
from flask_session import Session

app = Flask(__name__)
app.config['SESSION_TYPE'] = 'filesystem'  # Use server-side sessions
Session(app)

@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'POST':
        username = request.form['username']
        session['username'] = username
        return redirect(url_for('dashboard'))
    return render_template('index.html')

@app.route('/dashboard')
def dashboard():
    username = session.get('username')
    return render_template('dashboard.html', username=username)

@app.route('/logout')
def logout():
    session.pop('username', None)
    return redirect(url_for('index'))

if __name__ == '__main__':
    app.run(debug=True)
```

```python
Intermediate Flask Topics:
    6. Build a Flask app that allows users to upload files and display them on the website.
```

```python
flask_file_upload/
|-- app.py
|-- templates/
|   |-- index.html
|-- uploads/
```

```python
from flask import Flask, render_template, request, redirect, url_for, send_from_directory
import os

app = Flask(__name__)
app.config['UPLOAD_FOLDER'] = 'uploads'
app.config['ALLOWED_EXTENSIONS'] = {'txt', 'pdf', 'png', 'jpg', 'jpeg', 'gif'}

def allowed_file(filename):
    return '.' in filename and filename.rsplit('.', 1)[1].lower() in app.config['ALLOWED_EXTENSIONS']

@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'POST':
        # Check if the post request has the file part
        if 'file' not in request.files:
            return redirect(request.url)

        file = request.files['file']

        # If the user does not select a file, the browser submits an empty part without a filename
        if file.filename == '':
            return redirect(request.url)

        # If the file is allowed and the user submitted a non-empty file
        if file and allowed_file(file.filename):
            # Save the file to the uploads folder
            file.save(os.path.join(app.config['UPLOAD_FOLDER'], file.filename))
            return redirect(url_for('index'))

    # Display the list of uploaded files
    files = os.listdir(app.config['UPLOAD_FOLDER'])
    return render_template('index.html', files=files)

@app.route('/uploads/<filename>')
def uploaded_file(filename):
    return send_from_directory(app.config['UPLOAD_FOLDER'], filename)

if __name__ == '__main__':
    if not os.path.exists(app.config['UPLOAD_FOLDER']):
        os.makedirs(app.config['UPLOAD_FOLDER'])
    app.run(debug=True)
```

```python
Install Flask-SQLAlchemy:
Before you start, install the Flask-SQLAlchemy extension:


pip install Flask-SQLAlchemy
```

```python
from flask import Flask, render_template, request, redirect, url_for
from flask_sqlalchemy import SQLAlchemy
import os

app = Flask(__name__)
app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///db/app.db'
app.config['SQLALCHEMY_TRACK_MODIFICATIONS'] = False
db = SQLAlchemy(app)

class Item(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    name = db.Column(db.String(255), nullable=False)

def init_db():
    db.create_all()

@app.route('/')
def index():
    items = Item.query.all()
    return render_template('index.html', items=items)

@app.route('/add', methods=['POST'])
def add_item():
    name = request.form['name']
    new_item = Item(name=name)
    db.session.add(new_item)
    db.session.commit()
    return redirect(url_for('index'))

@app.route('/edit/<int:item_id>', methods=['GET', 'POST'])
def edit_item(item_id):
    item = Item.query.get(item_id)

    if request.method == 'POST':
        item.name = request.form['name']
        db.session.commit()
        return redirect(url_for('index'))

    return render_template('edit.html', item=item)

@app.route('/delete/<int:item_id>')
def delete_item(item_id):
    item = Item.query.get(item_id)
    db.session.delete(item)
    db.session.commit()
    return redirect(url_for('index'))

if __name__ == '__main__':
    if not os.path.exists('db/app.db'):
        init_db()
    app.run(debug=True)
```

```python
8. Implement user authentication and registration in a Flask app using Flask-Login.
```

```python
flask_user_auth/
|-- app.py
|-- templates/
|   |-- login.html
|   |-- register.html
|   |-- dashboard.html
|-- db/
|   |-- users.db
```

```python
Install Flask-Login:
Before you start, install the Flask-Login extension:

pip install Flask-Login
```

```python
from flask import Flask, render_template, request, redirect, url_for, flash
from flask_sqlalchemy import SQLAlchemy
from flask_login import LoginManager, UserMixin, login_user, login_required, logout_user, current_user
import os

app = Flask(__name__)
app.config['SECRET_KEY'] = 'secret_key'  # Change this to a strong secret key
app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///db/users.db'
app.config['SQLALCHEMY_TRACK_MODIFICATIONS'] = False
db = SQLAlchemy(app)
login_manager = LoginManager(app)
login_manager.login_view = 'login'

# User model
class User(UserMixin, db.Model):
    id = db.Column(db.Integer, primary_key=True)
    username = db.Column(db.String(50), unique=True, nullable=False)
    password = db.Column(db.String(50), nullable=False)

@login_manager.user_loader
def load_user(user_id):
    return User.query.get(int(user_id))

def init_db():
    db.create_all()

@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']
        user = User.query.filter_by(username=username).first()
        
        if user and user.password == password:
            login_user(user)
            flash('Login successful!', 'success')
            return redirect(url_for('dashboard'))
        else:
            flash('Login failed. Please check your username and password.', 'error')

    return render_template('login.html')

@app.route('/logout')
@login_required
def logout():
    logout_user()
    flash('Logout successful!', 'success')
    return redirect(url_for('login'))

@app.route('/dashboard')
@login_required
def dashboard():
    return render_template('dashboard.html', username=current_user.username)

@app.route('/register', methods=['GET', 'POST'])
def register():
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']
        existing_user = User.query.filter_by(username=username).first()

        if existing_user:
            flash('Username already exists. Please choose a different one.', 'error')
        else:
            new_user = User(username=username, password=password)
            db.session.add(new_user)
            db.session.commit()
            flash('Registration successful! You can now log in.', 'success')
            return redirect(url_for('login'))

    return render_template('register.html')

if __name__ == '__main__':
    if not os.path.exists('db/users.db'):
        init_db()
    app.run(debug=True)
```

```python
9. Create a RESTful API using Flask to perform CRUD operations on resources like books or movies.
```

```python
flask_restful_api/
|-- app.py
```

```python
Install Flask-RESTful:
Before you start, install the Flask-RESTful extension:
pip install Flask-RESTful
```

```python
from flask import Flask, request
from flask_restful import Resource, Api, reqparse

app = Flask(__name__)
api = Api(app)

books = [
    {"id": 1, "title": "The Great Gatsby", "author": "F. Scott Fitzgerald"},
    {"id": 2, "title": "To Kill a Mockingbird", "author": "Harper Lee"},
    {"id": 3, "title": "1984", "author": "George Orwell"},
]

parser = reqparse.RequestParser()
parser.add_argument("title", type=str, help="Title of the book")
parser.add_argument("author", type=str, help="Author of the book")

class BookListResource(Resource):
    def get(self):
        return {"books": books}

    def post(self):
        args = parser.parse_args()
        new_book = {
            "id": len(books) + 1,
            "title": args["title"],
            "author": args["author"],
        }
        books.append(new_book)
        return {"message": "Book added", "book": new_book}, 201

class BookResource(Resource):
    def get(self, book_id):
        book = next((item for item in books if item["id"] == book_id), None)
        if book:
            return {"book": book}
        return {"message": "Book not found"}, 404

    def put(self, book_id):
        args = parser.parse_args()
        book = next((item for item in books if item["id"] == book_id), None)
        if book:
            book["title"] = args["title"]
            book["author"] = args["author"]
            return {"message": "Book updated", "book": book}
        return {"message": "Book not found"}, 404

    def delete(self, book_id):
        global books
        books = [item for item in books if item["id"] != book_id]
        return {"message": "Book deleted"}

api.add_resource(BookListResource, "/books")
api.add_resource(BookResource, "/books/<int:book_id>")

if __name__ == "__main__":
    app.run(debug=True)
```

```python
Real-time Development:
    11. Create a real-time chat application using Flask-SocketIO.
```

```python
flask_socketio_chat/
|-- app.py
|-- templates/
|   |-- index.html
```

```python
Install Flask-SocketIO:
Before you start, install the Flask-SocketIO extension:


pip install flask-socketio
```

```python
from flask import Flask, render_template
from flask_socketio import SocketIO, emit

app = Flask(__name__)
app.config['SECRET_KEY'] = 'secret_key'  # Change this to a strong secret key
socketio = SocketIO(app)

@app.route('/')
def index():
    return render_template('index.html')

@socketio.on('message')
def handle_message(msg):
    print('Received message:', msg)
    emit('message', msg, broadcast=True)

if __name__ == '__main__':
    socketio.run(app, debug=True)
```

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta http-equiv="X-UA-Compatible" content="IE=edge">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Flask SocketIO Chat</title>
</head>
<body>
    <ul id="messages"></ul>
    <form id="form" action="">
        <input id="message_input" autocomplete="off" /><button>Send</button>
    </form>
    <script src="https://cdnjs.cloudflare.com/ajax/libs/socket.io/4.0.0/socket.io.js"></script>
    <script>
        var socket = io.connect('http://' + document.domain + ':' + location.port);

        socket.on('message', function(msg) {
            var ul = document.getElementById('messages');
            var li = document.createElement('li');
            li.appendChild(document.createTextNode(msg));
            ul.appendChild(li);
        });

        document.getElementById('form').onsubmit = function() {
            var input = document.getElementById('message_input');
            socket.emit('message', input.value);
            input.value = '';
            return false;
        };
    </script>
</body>
</html>
```

```python
Steps:
Save the above code into the respective files and directory structure.
Open a terminal and navigate to the directory containing app.py.
Run the Flask app using the command:
python app.py
Visit http://127.0.0.1:5000/ in your web browser.
Open multiple browser tabs or use different devices to simulate multiple users.
Enter a message in one tab, and you should see the message appear in real-time on all other tabs.
Explanation:
The @socketio.on('message') decorator is used to handle messages received from clients.
The emit('message', msg, broadcast=True) method sends the message to all connected clients in real-time.
The JavaScript code in the index.html file uses Socket.IO to establish a WebSocket connection with the server.
When a message is submitted via the form, it is sent to the server using socket.emit('message', input.value).
The server broadcasts the message to all connected clients, and each client updates its chat interface accordingly.
```

```python
12. Build a Flask app that updates data in real-time using WebSocket connections.
```

```python
flask_realtime_update/
|-- app.py
|-- templates/
|   |-- index.html
```

```python
Install Flask-SocketIO:
Before you start, install the Flask-SocketIO extension:


pip install flask-socketio
```

```python
from flask import Flask, render_template
from flask_socketio import SocketIO, emit

app = Flask(__name__)
app.config['SECRET_KEY'] = 'secret_key'  # Change this to a strong secret key
socketio = SocketIO(app)

# Initial counter value
counter = 0

@app.route('/')
def index():
    return render_template('index.html', counter=counter)

@socketio.on('increment')
def handle_increment():
    global counter
    counter += 1
    emit('update_counter', counter, broadcast=True)

if __name__ == '__main__':
    socketio.run(app, debug=True)
```

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta http-equiv="X-UA-Compatible" content="IE=edge">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Flask SocketIO Real-time Update</title>
</head>
<body>
    <h1>Real-time Counter: <span id="counter">{{ counter }}</span></h1>
    <button id="increment_button">Increment Counter</button>
    <script src="https://cdnjs.cloudflare.com/ajax/libs/socket.io/4.0.0/socket.io.js"></script>
    <script>
        var socket = io.connect('http://' + document.domain + ':' + location.port);

        // Update counter value on initial load
        document.getElementById('counter').textContent = {{ counter }};

        // Listen for counter updates
        socket.on('update_counter', function(counter) {
            document.getElementById('counter').textContent = counter;
        });

        // Increment counter on button click
        document.getElementById('increment_button').onclick = function() {
            socket.emit('increment');
        };
    </script>
</body>
</html>
```

```python
Steps:
Save the above code into the respective files and directory structure.
Open a terminal and navigate to the directory containing app.py.
Run the Flask app using the command:

python app.py
Visit http://127.0.0.1:5000/ in your web browser.
Open multiple browser tabs or use different devices to simulate multiple clients.
Click the "Increment Counter" button in one tab, and you should see the counter update in real-time on all other tabs.
Explanation:
The @socketio.on('increment') decorator is used to handle a custom event named 'increment' sent from clients.
The emit('update_counter', counter, broadcast=True) method sends the updated counter value to all connected clients in real-time.
The JavaScript code in the index.html file uses Socket.IO to establish a WebSocket connection with the server.
The counter value is initially loaded from the server, and subsequent updates are received in real-time.
```

```python
13. Implement notifications in a Flask app using websockets to notify users of updates.
```

```python
flask_notifications/
|-- app.py
|-- templates/
|   |-- index.html
```

```python
Install Flask-SocketIO:
Before you start, install the Flask-SocketIO extension:

pip install flask-socketio
```

```python
from flask import Flask, render_template, request
from flask_socketio import SocketIO, emit

app = Flask(__name__)
app.config['SECRET_KEY'] = 'secret_key'  # Change this to a strong secret key
socketio = SocketIO(app)

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/notify', methods=['POST'])
def notify():
    message = request.form.get('message', '')
    emit('notification', message, broadcast=True)
    return 'Notification sent'

if __name__ == '__main__':
    socketio.run(app, debug=True)
```

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta http-equiv="X-UA-Compatible" content="IE=edge">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Flask SocketIO Notifications</title>
</head>
<body>
    <div id="notifications"></div>
    <form id="notification_form" action="/notify" method="post">
        <input type="text" id="message_input" name="message" placeholder="Enter notification message" required />
        <button type="submit">Send Notification</button>
    </form>
    <script src="https://cdnjs.cloudflare.com/ajax/libs/socket.io/4.0.0/socket.io.js"></script>
    <script>
        var socket = io.connect('http://' + document.domain + ':' + location.port);

        socket.on('notification', function(message) {
            var div = document.createElement('div');
            div.appendChild(document.createTextNode(message));
            document.getElementById('notifications').appendChild(div);
        });

        document.getElementById('notification_form').onsubmit = function() {
            var input = document.getElementById('message_input');
            socket.emit('notification', input.value);
            input.value = '';
            return false;
        };
    </script>
</body>
</html>
```

```python
Steps:
Save the above code into the respective files and directory structure.
Open a terminal and navigate to the directory containing app.py.
Run the Flask app using the command:

python app.py
Visit http://127.0.0.1:5000/ in your web browser.
Enter a notification message in the input field and click "Send Notification."
The notification message should appear in real-time on all connected clients.
Explanation:
The /notify route allows clients to send notifications to the server by submitting a form.
The emit('notification', message, broadcast=True) method sends the notification message to all connected clients in real-time.
The JavaScript code in the index.html file uses Socket.IO to establish a WebSocket connection with the server.
Notifications are displayed in real-time as they are received from the server.
```
